# What is a Local Embedding Model?

A **local embedding model** is an embedding model that runs entirely on your own machine — no internet connection, no external API calls, no data leaving your system.

It converts text (words, sentences, documents) into **numerical vectors** that capture semantic meaning, just like cloud-based embedding services (e.g. OpenAI's `text-embedding-ada-002`), but everything stays local.

---

## What Are Embeddings, Exactly?

An embedding transforms text into a list of numbers (a vector) that represents its *meaning* in high-dimensional space. Texts with similar meanings end up close together in that space.

```
"I love cats"   →  [0.21, -0.45, 0.88, ...]
"I adore cats"  →  [0.23, -0.43, 0.85, ...]   ← very close (similar meaning)
"stock market"  →  [-0.67, 0.12, -0.34, ...]  ← far away (different meaning)
```

This is what powers semantic search and retrieval in RAG pipelines.

---

## Local vs Cloud Embedding

| | Local Model | Cloud API (e.g. OpenAI) |
|---|---|---|
| **Privacy** | Data never leaves your machine | Data sent to external server |
| **Cost** | Free after setup | Pay per token |
| **Speed** | Fast (no network latency) | Depends on internet |
| **Setup** | Requires install | Just an API key |
| **Offline** | Works offline | Requires internet |

## Popular Local Embedding Models

| Model | Size | Good For |
|---|---|---|

| `all-mpnet-base-v2` | ~420MB | Better quality, still lightweight |
| `nomic-embed-text` | ~270MB | Strong performance, via Ollama |
| `mxbai-embed-large` | ~670MB | High quality, via Ollama |
| `bge-m3` | ~570MB | Multilingual support |

## How to Use One in Python

**Option 1 — via `sentence-transformers`:**

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")  # downloads once, runs locally

texts = ["Machine learning is a subset of AI.", "I love pizza."]
embeddings = model.encode(texts)

print(embeddings.shape)  # (2, 384) — 2 texts, 384-dim vectors
```

**Option 2 — via Ollama (since you already have it):**

```bash
ollama pull nomic-embed-text
```

```python
import ollama

response = ollama.embeddings(
    model="nomic-embed-text",
    prompt="Machine learning is a subset of AI."
)
print(response["embedding"])  # list of floats
```

## Where It Fits in a RAG Pipeline

```
Your Documents
      ↓
  [Chunking]
      ↓
  [Local Embedding Model]  ← converts each chunk to a vector
      ↓
  [Vector Database]        ← stores vectors locally (e.g. ChromaDB, FAISS)
      ↓
  [Query comes in]
      ↓
  [Embed the query]        ← same local model
      ↓
  [Find closest chunks]    ← vector similarity search
      ↓
  [Local LLM]              ← e.g. Llama via Ollama
```

> **Tip:** Since you're already using Ollama for Llama, `nomic-embed-text` via Ollama is the most seamless way to add local embeddings — everything stays in one place.
